# Collaborative Filter Recommendation System

```python
books_rating=ratings.merge(books,on="ISBN")
```

For better results we are keeping users who have rated books more than 200 (in terms of frequency) from this we will get results from experienced users.

```python
books_rating_df=books_rating.groupby('User-ID').count()['Book-Rating']>200
books_rating_df[books_rating_df]
```

from the results we get that there are total number of 811 users who have given reviews on more than 200 books

```python
experienced_users=books_rating_df[books_rating_df].index
experienced_users
```

```python
books_rating["User-ID"].isin(experienced_users)
```

```python
filtered=books_rating[books_rating["User-ID"].isin(experienced_users)]
```

```python
#here we are finding popular books
```

```python
book_filtered=filtered.groupby('Book-Title').count()["Book-Rating"]>=50
popular_books=book_filtered[book_filtered].index
```

```python
popular_books
```

```python
filtered
```

Above we have the dataset of the experienced readers

```python
final=filtered[filtered["Book-Title"].isin(popular_books)]
```

```python
final
```

```python
table=final.pivot_table(index='Book-Title',columns="User-ID",values="Book-Rating")
```

```python
table.fillna(0,inplace=True)
```

```python
table
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn_model = NearestNeighbors(n_neighbors=7, metric='cosine', algorithm='brute')
knn_model.fit(table)
```

```python
def book_ML(name):
    recommendation = []
    book_index = np.where(table.index == name)[0][0]
    distances, indices = knn_model.kneighbors([table.iloc[book_index]], n_neighbors=10)
    for index in indices[0][1:]:
        recommendation.append(table.index[index])
    return recommendation
```

```python
book_ML("1984")
```

```python
book_ML("The Handmaid's Tale")
```

```python
book_ML("Brave New World")
```

```python
# Assuming 'table' is your data
distances, indices = knn_model.kneighbors(table, n_neighbors=7)
avg_similarity = np.mean(distances)
print("Average Similarity of Recommendations:", avg_similarity)
```
